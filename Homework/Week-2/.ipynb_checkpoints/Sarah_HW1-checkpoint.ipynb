## 1. Data Types and Summary
*Identify data type for each variable in the  movie dataset and provide a summary of these variables*
+ **Genre**: String/object (categorical), describing genre of the movie
+ **Rotten Tomatoes**: Integer, giving a rating from 0 to 100. We have one invalid entry (125).
+ **IMDB**: Float, giving a rating from 1 to 10.
+ **Audience Ratings**: Integer, the amount of people who watched and rated the movie
+ **Budget_M**: Float, the budget for the movie in millions.
+ **Sales Revenue_M**: Float, the amount of money the movie makes in millions
+ **Year**: When checking, the datatype is put as an object. This is a discrete numerical variable, telling the year the movie came out

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('movie.csv')
df
```

```python
df.dtypes
```

## 2. Descriptive Statistics
*What are the descriptive statistics (mean, median, standard deviation, etc.) for each numerical column?*

```python
for col in df.select_dtypes(include='number').columns:
    print(f"{col}\n{df[col].describe()}\n")
```

```python
for col in df.select_dtypes(include='number').columns:
    print(f"{col} median:\n{df[col].median()}\n")
```

## 3. Distribution Analysis
*Create a histogram and boxplots for each independent variable to examine their distribution and skewness. Indicate which variables exhibit right/left skewness or normal distribution*
+ **Genre**: Normal
+ **Rotten Tomatoes**: Right skewed
+ **IMDB**: Normal
+ **Audience Ratings**: Normal/left skeweed (little bit)
+ **Budget_M**: Right skewed (Extreme)
+ **Year**: Normal

```python
indep = ['Genre', 'Rotten Tomatoes', 'IMDB', 'Audience Ratings','Budget_M', 'Year of Release']
indep_data = df[indep]

for col in indep_data.columns:
    sns.histplot(indep_data[col], kde=True)
    plt.title(f"Histogram and Density Plot of {col}")
    plt.show()
```

## 10. Noisy Data Analysis
*Examine the dataset for noisy data, including errors and imprecise entries. How might these affect the analysis?*
+ There's 2 different columns for SciFi, since people spelled it differently
+ Budget_M has some crazy numbers, over 20,000,000,000. No movie can make that much, we can delete those higher numbers.

```python
for col in df.select_dtypes(include='object').columns:
    print(f"Value of {col}: \n{df[col].value_counts()}\n")
```

```python
invalid_entries = df[df['Rotten Tomatoes']>100].count()
print("Invalid Entries:\n", invalid_entries)
```

```python
sns.boxplot(x=df['Budget_M'])
plt.title(f"Box Plot of Budget_M")
plt.show()
```
