## 7. Outlier Detection with Boxplots
Create boxplots for each variable to visually identify outliers. How will the presence of outliers impact your analysis and how do you plan to address them?

```python
for col in df.select_dtypes(include=['number']).columns:
  sns.boxplot(x=df[col])
  plt.title(f"Box Plot of {col}")
  plt.show()
```

```python
for col in df.select_dtypes(include=['number']).columns:
    Q1 = df[col].describe()['25%']
    Q3 = df[col].describe()['75%']
    IQR = Q3-Q1
    out = IQR*1.5

    Q1_outlier = Q1 - out
    Q3_outlier = Q3 + out

    df.loc[df[col]>Q3_outlier,col] = None
    df.loc[df[col]<Q1_outlier,col] = None
```

```python
for col in df.select_dtypes(include=['number']).columns:
  sns.boxplot(x=df[col])
  plt.title(f"Box Plot of {col}")
  plt.show()
```

I used IQR*1.5 method to get ride of outliers. Having those outliers would skew my results, leading to an inaccurate result

## 10. Outlier Detection with Z-Score
Use the Z-Score method to detect outliers in the dataset. How will you decide which outliers to handle and what strategies will you employ?

+ I would make the threshold z-score be 3 and -3, then either replace them with the values at those max or min z-score, or just remove them

```python
for col in df.select_dtypes(include=['number']).columns:
    mean = df[col].mean()
    std = df[col].std()

    zscores = (df[col] - mean) / std

    threshold = 3

    outliers = (df[abs(zscores) > threshold])

    print(f"Outliers in {col} detected using z-score method:\n", outliers, "\n")
```
